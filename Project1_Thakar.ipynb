```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
import numpy 
from scipy.stats import ttest_ind
%matplotlib inline


workbook1 = pd.read_excel('alz12089-sup-0001-suppmat.xlsx', 'Supplementary table 2', skiprows=[0],nrows=18)
workbook2 = pd.read_excel('alz12089-sup-0001-suppmat.xlsx', 'Supplementary table 2', skiprows=20,nrows=20)
#print(workbook1)  # print workbook or use .head to see first 5 rows

#print(workbook1.shape) #print the shape of the dataframe
#workbook.columns #make sure the names of the columns are proper

```

```python
CTRL_data = workbook1.iloc[0:18,0:8] #Establish CTRL dataset
AD_data = workbook2.iloc[0:20,0:8]  #Establish AD dataset
#print(df1_data.shape)
#print('rows:\n', df1_data)
#print(AD_data)
```

```python
sns.set(style='whitegrid')

t_tau_CTRL = workbook1.iloc[0:18,3] #identifying location of t-tau protein in CTRL and AD patients
t_tau_AD = workbook2.iloc[0:21,3]
both = [t_tau_CTRL, t_tau_AD] #putting both datasets in one list


#plot1 = sns.boxplot(data=both, whis=0.25,showfliers=False, showbox=False)
#plot1 = sns.swarmplot(data=both, dodge=True, size =6) #practice swarmplot 

#plot2 = sns.boxplot(y=t_tau_AD, showbox=False)
#plot1 = plt.boxplot(x=df1_data[0:18,3], sym='c', showbox=False)
```

```python
both1 = pd.concat((CTRL_data,AD_data)) #work off this combined dataset

plot = sns.boxplot(x='Disease',y='total tau/EV (ng/mg)', data=both1, whis=0.25,showfliers=False, showbox=False)
plot = sns.swarmplot(x='Disease',y='total tau/EV (ng/mg)', data=both1, dodge=True, size =6)
plot.set(ylabel='t-tau/EV protein (ng/mg)')
plt.ylim(0,4) #adjust y range 
plt.title('t-tau',fontsize=20) #adjust title and size
plot.grid(False) #turn off gridlines

fig = plot.get_figure()
fig.savefig('t-tau.png') # save figure


ttest_ind(t_tau_CTRL, t_tau_AD)
```
